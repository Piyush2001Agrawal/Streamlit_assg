```python
avg_amt = round(df['Amount'].mean())
print(f"Average Amount: {avg_amt}")
```

```python
df['Amount'].fillna(avg_amt, inplace=True)
df.isnull().sum()
```

```python
avg_amt_per_ord = round(df['Avg Amount per Order'].mean())
print(f"Average Amount Per Order: {avg_amt_per_ord}")
```

```python
df['Avg Amount per Order'].fillna(avg_amt_per_ord, inplace=True)
df.isnull().sum()
```

```python
cols_to_drop =["Status","unnamed1"]
df= df.drop(columns = cols_to_drop)
df.head()
```

Diwali Sales Data Visualization using Seaborn

```python
sns.histplot(df['Amount'], kde=True,bins=30,color='Violet')
plt.title('Distribution of Purchase Amount')
plt.show()
```

```python
sns.barplot(data= df,x='Gender',y='Amount',hue='Age Group',palette='Set2',estimator=sum)
plt.title("Total Amount by Gender and Age Group")
```

```python
sns.countplot(x='Age Group', data=df, palette='viridis')
plt.title('Age Group Distribution')
plt.show()
```

```python
sns.countplot(data= df, x='Gender', palette='Set2')
plt.title('Gender Count')
plt.show()
```

```python
sns.boxplot(x='Gender', y='Amount', data=df,palette='viridis')
plt.title('Amount Distribution by Gender')
plt.show()
```

```python
sns.violinplot(x='Marital_Status', y='Amount', data=df, palette='muted')
plt.title('Amount Distribution by Marital Status')
plt.show()
```

```python
sns.countplot(data=df,y='Product_Category', order=df['Product_Category'].value_counts().index, palette='cubehelix')
plt.title('Product Category Distribution')
plt.show()
```

```python
sns.heatmap(df[['Orders', 'Amount', 'Age']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Orders', 'Amount', 'Age']], diag_kind='kde', palette='viridis')
plt.title('Pairplot of Orders, Amount, and Age', fontsize=16,y=1.02)
plt.show()
```

```python
sns.lineplot(x='Age Group', y='Amount', data=df,  marker='o')
plt.title('Total Amount by Age Group')
plt.show()
```

```python
sns.stripplot(data=df, x='Age Group', y="Amount", jitter=True, hue='Gender', palette='Set1')
plt.title('Amount Distribution by Age Group')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper right')
plt.show()
```

Diwali Sales Data Visualization using Plotly

```python
px.bar(df,x='Gender', y='Amount', color='Product_Category', barmode='group',
       title='Total Amount by Gender')
```

```python
px.bar(df, x='Age Group', y='Orders', color='Age Group', 
       title='Total Orders by Age Group')
```

```python
px.pie(df, names='Gender',
       title='Gender Distribution',template='plotly_dark',)
```

```python
px.scatter(df, x='Orders', y='Amount', color='Gender', 
           title='Scatter Plot of Orders vs Amount',)
```

```python
px.box(df, x='Gender', y='Amount', color='Gender', 
       title='Amount Distribution by Gender')
```

```python
top_states = df.groupby('State')['Amount'].sum().nlargest(10).reset_index()
px.bar(top_states, x='State', y='Amount', color='State', title='Top 10 States by Total Amount',template='plotly_dark')
```

```python
px.sunburst(df, path=['Gender', 'Age Group'], values='Amount', 
             title='Amount by Gender and Age Group',template='plotly_dark')
```

```python
px.scatter(df, x='Age', y='Amount', color='Gender', 
           title='Age vs Amount')
```

```python
px.imshow(df[['Orders', 'Amount', 'Age']].corr(), text_auto=True, 
          title='Correlation Heatmap',template='plotly_dark',
          color_continuous_scale='YlGnBu')
```

```python
px.funnel(df.groupby('Age Group')['Amount'].sum().reset_index(), x='Amount', y='Age Group',
          title='Funnel Chart of Amount by Age Group',template='plotly_dark')
```

```python
px.violin(df, x='Product_Category', y='Orders', color='Product_Category',
       title='Total Orders by Product Category',template='plotly_dark',
       color_discrete_sequence=px.colors.qualitative.Set2)
```

```python
px.line(df, x='Gender',y ='Orders', markers='o',
        title='Total Orders by Gender',template='plotly_dark',color_discrete_sequence=px.colors.qualitative.Set3)
```

```python
px.treemap(df, path=['State', 'Product_Category'], values='Orders',
           title='Treemap of Orders by State and Product Category',template='plotly_dark')
```

```python
px.scatter(df, x='Age', y='Orders', size='Amount', color='Zone',
                 title='Bubble Chart: Orders vs Age (Bubble Size = Amount)',
                 hover_data=['State', 'Product_Category'])
```
